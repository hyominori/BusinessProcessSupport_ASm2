```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv (r'C:\Users\Admin\Documents\Zalo Received Files\asm2_data.csv')
data
```

Cleaning data

```python
data.isnull().sum() # Cho thấy những dòng nào bị trống
```

```python
data.dropna(inplace=True) # Loại bỏ những dòng trống
data.isnull().sum() # Kiểm tra lại
```

```python
duplicates = data[data.duplicated(keep=False)] # Cho thấy những dòng bị trùng
print(duplicates)
```

```python
data.drop_duplicates(inplace=True) # Xóa những hàng trùng lặp
data # Xem lại dữ liệu
```

```python
data.info() # Xem thông tin của data
```

Chage data Type

```python
data['ProductID'] = data['ProductID'].astype('int64') # đổi data type của 'ProductID' từ float64 thành int64
data['Quantity Ordered'] = data['Quantity Ordered'].astype('int64') # đổi data type của 'Quantity Ordered' từ float64 thành int64
data['Date'] = pd.to_datetime(data['Date']) # đổi data type của 'Date' từ object thành datetime
```

```python
data.info()
```

Replace Data

```python
data.replace('NYC', 'New York City', inplace = True)
data.replace('LA', 'Los Angeles', inplace = True)
```

Merge data

```python
data['Detailed Address'] = data['Address'] + ' ' + data['City']
data
```

Create a new column

```python
data['Sale'] = data['Price Each'] * data['Quantity Ordered']
data
```

```python
data['Month'] = data['Date'].dt.month #  Tạo một cột mới 'Month' phục vụ cho việc tạo biểu đồ
```

```python
monthly_sales = data.groupby('Month')['Sale'].sum() #Tính tổng Sale theo Month
```

```python
# Vẽ biểu đồ đường
plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')

# Thêm doanh thu trên từng điểm
for i in range(len(monthly_sales)):
    plt.annotate(f'{monthly_sales.values[i]:.2f}', 
                 (monthly_sales.index[i], monthly_sales.values[i]), 
                 textcoords="offset points", 
                 xytext=(0,10), 
                 ha='center')
    
plt.plot(monthly_sales.index, monthly_sales.values)
plt.title('Total Sales per Month')
plt.xlabel('Month')
plt.ylabel('Sales per Month')
plt.show()
```

```python
city_sales = data.groupby('City')['Sale'].sum().reset_index() # Tính tổng doanh thu theo thành phố
```

```python
# Hàm để định dạng phần trăm và doanh thu
def func(pct, all_values):
    total = sum(all_values)
    val = pct / 100. * total
    return f'{val:.2f}\n({pct:.2f}%)'

# Vẽ biểu đồ tròn
plt.figure(figsize=(10, 8))
plt.pie(city_sales['Sale'], labels=city_sales['City'], autopct=lambda pct: func(pct, city_sales['Sale']),
        startangle=140, colors=plt.cm.Paired.colors)
plt.title('Total Sales by City')
plt.axis('equal')  # Đảm bảo biểu đồ tròn là hình tròn
plt.show()
```

```python
# Tính tổng doanh thu theo StoreID
store_sales = data.groupby('StoreID')['Sale'].sum().reset_index()
```

```python
# 10 Highest Sales by StoreID
top_stores = store_sales.sort_values('Sale', ascending=False).head(10)
print("Top 10 Highest Sales by Store")
print(top_stores)
```

```python
# 10 Lowest Sales by StoreID
bottom_stores = store_sales.sort_values('Sale', ascending=True).head(10)
print("Top 10 Lowest Sales by Store")
print(bottom_stores)
```
